# #0 The Harmonic Oscillator

Everything always comes back to the Harmonic Oscillator. Here we will review the classical and quantum harmonic oscillator.

## The Classical Oscillator

The 1D classical harmonic oscillator is defined by the potential
$$ V(x) = \frac{1}{2}kx^2,$$
where $k$ is some constant. Such a potential appears in many different areas of Physics that exhibit oscillatory motion. The most common example being a mass on a spring, where there is some kind of linear restoring force acting on a mass. One can see this by deriving the force resulting from this potential:
$$ \begin{split} \overrightarrow{F} &= -\overrightarrow{\nabla}V(x) \\ &= -\frac{d}{dx}\left[ \frac{1}{2}kx^2\right]\\ &= -kx \end{split}$$
where even though our potential is only 1 dimensional, I have used vector notation to keep the discussion general to higher dimensions.

We can derive the equations of motion resulting from this potential from Newton's second law:
$$ \frac{d^2 x(t)}{dt^2} = -\frac{k}{m}x(t),$$
which gives the general solution
$$ x(t) = A\sin(\omega t) + B\cos(\omega t),$$
with $\omega = \sqrt{\frac{k}{m}}$. Provided with initial conditions, one can determine the coefficients $A$ and $B$ and know the full trajectory of the particle. I'm sure this won't blow anyone's mind.

Instead, let's look at the other formulations of classical mechanics to see what other insights they can offer about the classical oscillator. We can define a Lagrangian:
$$ L = T - V $$
with $T = \frac{1}{2}m\dot{x}^2$ being the kinetic energy (using the dot notation for time derivatives), and $V$ our potential function. Our Lagrangian for the oscillator is therefore:
$$ L = \frac{1}{2}m\dot{x}^2 - \frac{1}{2}kx^2.$$
The Euler-Lagrange equations,
$$ \frac{\partial L}{\partial x} - \frac{d}{dt}\left( \frac{\partial L}{\partial \dot{x}}\right) = 0$$
allow us to determine the equation of motion,
$$ -kx - \frac{d}{dt}\left( m\dot{x}\right) = 0$$
$$ \ddot{x} = -\frac{k}{m}x, $$
the same equation we obtained before.

Finally, let's transform the Lagrangian to a Hamiltonian using the Legendre transform,
$$ H = p\dot{q} - L,$$
where $p$ is the conjugate momenta of the $q$ coordinate,
$$ p = \frac{\partial L}{\partial \dot{q}}.$$
For our coordinate $x$, it's conjugate momentum is
$$ p = \frac{\partial L}{\partial \dot{x}} = m\dot{x},$$
i.e. the classical momentum. Re-writing our coordinate $\dot{x}$ in terms of the momentum $p$, $\dot{x}=p/m$, we obtain the Hamiltonian,
$$ H = \frac{p^2}{2m} + \frac{1}{2}kx^2.$$
One can easily see this is the total energy of the system. Now, we already know what the classical solution to the equation of motion is,
$$ x(t) = A\sin(\omega t) + B\cos(\omega t),$$
with $\omega = \sqrt{\frac{k}{m}}$ and substituting into the Hamiltonian, we do indeed find it is a constant,
$$ H = \frac{m\omega^2}{2}(A^2 + B^2)=E,$$
which is good, as this confirms that energy is conserved. If we treat $p$ and $x$ as independent coordinates, we note that the Hamiltonian simply defines the equation of an ellipse in the $xp$ plane, and so every energy $E$ defines an elliptical trajectory in phase space.

Before looking at the quantum harmonic oscillator, I want to take a brief side-step to discuss using Hamilton's equations of motion to solve for the particle trajectory. Hamilton's equations of motion are
$$ \dot{x} = \frac{\partial H}{\partial p},$$
$$ \dot{p} = -\frac{\partial H}{\partial x}. $$
So, instead of solving a second order differential equation, we instead of two (coupled) first order differential equations. Substituting our Hamiltonian, we have
$$ \dot{x} = \frac{p}{m},$$
$$ \dot{p} = -kx = -m\omega^2 x.$$
I'm not going to use the constant $k$ in future, instead I will use the natural frequency $\omega$. We can define a matrix equation,
$$ \begin{pmatrix} \dot{x} \\ \dot{p} \end{pmatrix} = \begin{pmatrix} 0 & \frac{1}{m} \\ -m\omega^2 & 0 \end{pmatrix} \begin{pmatrix} x \\ p \end{pmatrix},$$
which captures our coupled differential equations. To solve, we want to diagonalise our matrix by finding its eigenvalues and eigenvectors. The eigenvalue equation is given by
$$ M\vec{v} = \lambda \vec{v},$$
where for us, our matrix $M$ is
$$ M = \begin{pmatrix} 0 & \frac{1}{m} \\ -m\omega^2 & 0 \end{pmatrix}. $$
First, to find the eigenvalues, we look for $\lambda$ such that $det(M-\lambda I) = 0$, i.e. the roots of the matrix's characteristic polynomial. This is simply
$$ \lambda^2 + \omega^2 = 0,$$
giving the eigenvalues
$$ \lambda_{\pm} = \pm i\omega.$$
The corresponding eigenvectors are then
$$ \vec{v}_{\pm} = \begin{pmatrix} 1 \\ \pm im\omega \end{pmatrix},$$
and hence, the corresponding change of basis matrix is
$$ P = \begin{pmatrix} 1 & 1 \\ im\omega & -im\omega \end{pmatrix}.$$
Applying this to our state, we arrive at the new coordinates
$$ \begin{pmatrix} a_+ \\ a_- \end{pmatrix} = P^{-1}\begin{pmatrix} x \\ p\end{pmatrix} = \frac{1}{2}\begin{pmatrix} x - \frac{ip}{m\omega} \\ x + \frac{ip}{m\omega} \end{pmatrix}.$$
